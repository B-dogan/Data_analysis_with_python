#### aggregation & grouping

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df["age"].mean()
```

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("sex")[["age"]].mean()
```

## birden fazla aggregation'ı nasıl uygularım 

```python
df.groupby("sex").agg({"age":np.mean})   #### bunu tek bir agg için bile kullanmaya alışalım ki rahat edelim ileride
```

```python
df.groupby("sex").agg({"age":[np.mean,"sum"]}) 
```

```python
df.groupby("sex").agg({"age":[np.mean,"sum"],
                       "embark_town":"count" }) 
```

```python
### bana verdiği şey embarkın frekansı değil cinsiyetin frekansı oldu !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```

```python
df.groupby("sex").agg({"age":[np.mean,"sum"],
                       "survived":"mean"}) 
```

```python
df.groupby(["sex","embark_town"]).agg({"age":[np.mean,"sum"],                   ### groupby seviyelerini artırdık.
                       "survived":"mean"}) 
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":[np.mean,"sum"],            ### 3 kırılımda.     
                       "survived":"mean"}) 
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":[np.mean,"sum"],           
                       "survived":"mean",
                        "sex":"count"}) 
```
