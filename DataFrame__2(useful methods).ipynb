## condition yazarken loc ile yazmaya özen gösterelim hata almamak veya istediğimizi rahatlıkla alabilmek için

df.loc[()     &   ()  ]    böyle bir şey gördüğümde ben anlamalıyım ki 2 condition var satırlar üzerinde bakacağız sadece çünkü neden : virgül yok  ayrıca parantez kullanmaz isek hata alırız veya istediğimizi alamayız.

```python
df.loc[(df["sepal_length"] > 4) & (df["sepal_length"] <5 ) ] 
```

sepal lengthi 4 ie 5 arasındaki çiçekler genelde setosaymış gibi bir çıkarım yapabiliriz

diyelim ki bu condition'u yazdım ve sadece virginicaya ulaşmak istiyorum:

```python
df.loc[(df["sepal_length"] > 4) & (df["sepal_length"] <5 ) & (df["species"] == "virginica")]
```

parantezlere dikkat edelim & | vs varsa parantez kullanalım yoksa istediğimiz sonuçları alamayabiliriz.

```python
df.loc[(df["sepal_length"] > 4) & (df["sepal_length"] <5 ) & (df["species"] == "virginica"),"sepal_length"]
```

```python
df.loc[(df["sepal_length"] > 4) & (df["sepal_length"] <5 ) & (df["species"] == "virginica"),["sepal_length"]]
```
